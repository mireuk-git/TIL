# Preprocessing

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

titanic = sns.load_dataset('titanic')
titanic
```

```python
titanic = titanic[['survived','pclass','sex','age','sibsp','parch','fare','embarked']].dropna()
titanic['sex']=titanic['sex'].map({'male':0,'female':1})
titanic['embarked']=titanic['embarked'].map({'C':0,'Q':1,'S':2})
titanic
```

```python
x=titanic.drop('survived',axis=1)
y=titanic['survived']
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)

scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

# Logistic

```python
from sklearn.linear_model import LogisticRegression

LRmodel=LogisticRegression()
LRmodel.fit(x_train, y_train)

LR_pred = LRmodel.predict(x_test)
```

# SVM

```python
from sklearn.svm import SVC

SVMmodel = SVC()
SVMmodel.fit(x_train, y_train)

SVC_pred=SVMmodel.predict(x_test)
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier

KNNmodel = KNeighborsClassifier(n_neighbors=5)
KNNmodel.fit(x_train, y_train)
KNN_pred = KNNmodel.predict(x_test)
```

# Naive_bayes

```python
from sklearn.naive_bayes import GaussianNB

NBmodel = GaussianNB()
NBmodel.fit(x_train, y_train)

NB_pred=NBmodel.predict(x_test)
```

# Decision_Tree

```python
from sklearn.tree import DecisionTreeClassifier

treemodel = DecisionTreeClassifier()
treemodel.fit(x_train,y_train)

tree_pred = treemodel.predict(x_test)
```

# Evaluate 

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

model=["LogisticRegression","SVM","KNN","Naive_bayes","DicisionTree"]
pred=[LR_pred, SVC_pred, KNN_pred, NB_pred, tree_pred]
for i in range(len(model)):
    print(model[i])
    print("accuracy_score:",accuracy_score(y_test,pred[i]))
    print("classification_report:\n",classification_report(y_test,pred[i]))
    print("confusion_matrix:\n",confusion_matrix(y_test,pred[i]))
    print("\n\n\n")
```
